```python
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].median())
```

```python
titanic_df['Embarked']=titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode())
```

```python
titanic_df.drop(columns=['Cabin'], inplace=True)
```

```python
titanic_df['Sex'] = titanic_df['Sex'].map({'male': 0, 'female': 1})
titanic_df['Embarked'] = titanic_df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

Exploratory Data Analysis (EDA)


```python
numeric_df = titanic_df.drop(columns=['Name', 'Ticket'])
```

```python
titanic_df.hist(bins=10, figsize=(16, 7))
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(5, 4))
sns.barplot(x='Sex', y='Survived', data=titanic_df)
plt.title('Survival Rate by Sex')
plt.show()
```

```python
plt.figure(figsize=(5, 4))
sns.barplot(x='Pclass', y='Survived', data=titanic_df)
plt.title('Survival Rate by Class')
plt.show()
```

```python
plt.figure(figsize=(5, 4))
sns.histplot(data=titanic_df, x='Age', hue='Survived', multiple='stack')
plt.title('Survival Rate by Age')
plt.show()
```

```python
plt.figure(figsize=(5, 4))
sns.barplot(x='Embarked', y='Survived', data=titanic_df)
plt.title('Survival Rate by Embarked')
plt.show()
```

```python
sns.pairplot(titanic_df, hue='Survived', diag_kind='kde')
plt.show()
```
